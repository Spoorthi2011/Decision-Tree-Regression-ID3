# Data preprocessing

```python
df.isnull().sum().sort_values(ascending=False)       # Checking the null(missing) values
```

```python
# Replace missing values with the median value of the feature
df.fillna(df.median(), inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

# Implement Decision Tree Regression(ID3)

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

# Set the features and target variable
X =df.iloc[:, :-1]
y =df.iloc[:, -1]

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Initialize the decision tree regressor
dt_regressor = DecisionTreeRegressor(random_state=0)

# Fit the model on the training data
dt_regressor.fit(X_train, y_train)

# Make predictions on the test data
y_pred = dt_regressor.predict(X_test)

# Calculate MAE
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)

# Evaluate the model using mean squared error
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error:', mse)
```
